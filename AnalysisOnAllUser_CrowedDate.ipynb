```python
import numpy as np
import pandas as pd
```

```python
col_list = ['User', 'Latitude','Longitude','Date']
```

```python
df = pd.read_csv('all_trajectory.csv',usecols=col_list)
```

```python
df
```

```python
UserAndDate=df.drop(['Latitude','Longitude'],axis=1)
UserAndDate
```

Group user as Date

```python
UserDateGroupByDate = UserAndDate.groupby('Date')['User'].unique()
UserDateGroupByDate
```

```python
FrequencyUserByDate = UserAndDate.groupby('Date')['User'].nunique()
FrequencyUserByDate
```

```python
FrequencyUserByDate=FrequencyUserByDate.sort_values(axis=0)
FrequencyUserByDate
```

```python
MostUserDate=FrequencyUserByDate[FrequencyUserByDate==FrequencyUserByDate.max(axis=0)]
MostUserDate
```

Which date show most users?

```python
CrowedDate=str(MostUserDate.index[0])
CrowedDate
```

```python
UserInCrowedDate = UserDateGroupByDate[CrowedDate]
print(UserInCrowedDate)
```

Find user location in crowded date

```python
LocationInCrowedDate = df.loc[df['Date']==CrowedDate]
LocationInCrowedDate
```

Only pick the first Location of each user in Crowed Date

```python
FirstLocation_OfEachUser_InCrowedDate = LocationInCrowedDate.groupby('User').first().drop(['Date'],axis=1)
FirstLocation_OfEachUser_InCrowedDate
```

Export Location to CSV

```python
FirstLocation_OfEachUser_InCrowedDate.to_csv('FirstLocation_OfEachUser_InCrowedDate.csv')
```
